```python
import easyocr
import matplotlib.pyplot as plt
import cv2
import warnings

warnings.filterwarnings("ignore")
```

```python
# Create an EasyOCR Reader object
reader = easyocr.Reader(['en'])  # Specify language(s)
```

```python
# Load an image using OpenCV
image_path = 'S:\\Machine Learning\\Siddharthan Playlist\\Projects\\E_KYC_USING_COMPUTER_VISION\\WhatsApp Image 2024-07-30 at 19.32.08_eba7bab2.jpg'
image = cv2.imread(image_path)
```

```python
# Perform OCR on the image
results = reader.readtext(image_path)
```

```python
# Display the results
for result in results:
    # result is a tuple (bbox, text, confidence)
    bbox, text, confidence = result
    print(f"Detected text: '{text}' with confidence: {confidence}")
```

```python
# Plot the image with detected text
for bbox, text, confidence in results:
    # Draw bounding boxes and text
    points = [tuple(point) for point in bbox]
    for i in range(len(points)):
        cv2.line(image, points[i], points[(i+1) % len(points)], (0, 255, 0), 2)
    cv2.putText(image, text, (int(bbox[0][0]), int(bbox[0][1] - 10)),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
```

```python
# Convert BGR image to RGB for displaying with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Show the image with bounding boxes and text
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
