```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 as cv 
from deepface import DeepFace 
```

```python
cascade_path = 'S:\\Machine Learning\\Siddharthan Playlist\\Projects\\E_KYC_USING_COMPUTER_VISION\\haarcascade_frontalface_default.xml'
img_path = 'S:\\Machine Learning\\Siddharthan Playlist\\Projects\\E_KYC_USING_COMPUTER_VISION\\contour_id.png'

BGR_img = cv.imread(img_path) 
RGB_img = cv.cvtColor(BGR_img,cv.COLOR_BGR2RGB) 
gray_img = cv.cvtColor(BGR_img,cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(BGR_img) 
plt.axis("off")
plt.show() 
```

```python
plt.imshow(RGB_img) 
plt.axis("off")
plt.show() 
```

```python
plt.imshow(gray_img,cmap = 'gray') 
plt.axis("off")
plt.show() 
```

```python
face_cascade = cv.CascadeClassifier(cascade_path) 
face = face_cascade.detectMultiScale(gray_img,scaleFactor=  1.1, minNeighbors= 5) 
```

```python
face
```

```python
x, y, w, h = face[0]

new_w = int(w * 2) 
new_h = int(h * 2) 

new_x = max(0,x - int((new_w-w)/2))
new_y = max(0,y-int((new_h-h)/2)) 

extract_face = RGB_img[new_y : new_y + new_h, new_x : new_x + new_w] 
```

```python
plt.imshow(extract_face)  
plt.axis("off")
plt.show() 
```
